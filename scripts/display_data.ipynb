```python
from haven import haven_utils as hu
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math

datadir_base = '/home/xhdeng/scratch/datasets/cassava'

labels = hu.load_json(os.path.join(datadir_base, 'label_num_to_disease_map.json'))
train = pd.read_csv(os.path.join(datadir_base, 'train.csv'))
num_images = 9

image_label_dict = {'0': [], '1': [], '2': [], '3': [], '4': []}

for image_id, label in train.values:
    image_label_dict[str(label)].append(image_id)


print('%d images from each class' % num_images)
for label in image_label_dict.keys():
    fig = plt.figure()
    plt.title('class: ' + label)
    plt.axis('off')
    for index in range(1, num_images + 1):
        img = mpimg.imread(os.path.join(datadir_base, 'train_images', image_label_dict[label][index]))
        fig.add_subplot(math.ceil(math.sqrt(num_images)), int(math.sqrt(num_images)), index)
        plt.imshow(img)
        plt.axis('off')
    
    
```
